```python
import re

def process_raga_entry(raga_entry):
    raga_parts = raga_entry.split(maxsplit=1)
    raga_name = raga_parts[0]
    notes = raga_parts[1] if len(raga_parts) > 1 else ''

    split_notes = re.split(r'\s*-\s*', notes)

    arohanam = split_notes[0].split()
    avarohanam = split_notes[1].split() if len(split_notes) > 1 else []

    return [raga_name, arohanam, avarohanam]
```

```python
processed_ragas = [process_raga_entry(raga) for raga in ragas]

# Print the processed ragas
for processed_raga in processed_ragas:
    print(processed_raga)
```

```python
process_janya_janaka = []

for raga in processed_ragas:
    first_element = raga[0]
    if first_element.isdigit():
        raga_name = raga[1][0]
        arohanam = raga[1][1:]
        avarohanam = raga[2]
        process_janya_janaka.append([raga_name, arohanam, avarohanam, 'Janaka'])
    else:
        full = list(raga)
        full.append('Janya')
        process_janya_janaka.append(full)


process_janya_janaka
```

```python
import pandas as pd
from collections import Counter

def create_raga_dataset(df):
    notes = ['S', 'R1', 'R2', 'R3', 'G1', 'G2', 'G3', 'M1', 'M2',
             'D1', 'D2', 'D3', 'N1', 'N2', 'N3']

    processed_data = []

    for index, row in df.iterrows():
        raga_name = row['Raga']
        aarohanam = row['Aarohanam']
        avarohanam = row['Avarohanam']
        janya_janaka = row['Janya/Janaka']

        aarohanam_count = Counter(aarohanam)
        avarohanam_count = Counter(avarohanam)

        aarohanam_data = [aarohanam_count.get(note, 0) for note in notes]
        avarohanam_data = [avarohanam_count.get(note, 0) for note in notes]

        processed_data.append([raga_name, aarohanam, avarohanam, janya_janaka] + aarohanam_data + avarohanam_data)

    columns = ['Raga','Aarohanam','Avarohanam', 'Janya/Janaka'] + [note + '_Aarohanam' for note in notes] + [note + '_Avarohanam' for note in notes]

    new_df = pd.DataFrame(processed_data, columns=columns)

    return new_df

new_df = create_raga_dataset(df)
new_df.head()
```

```python
def calculate_base_count(base_tuple):
    base_count = {}

    for ragam in base_tuple:
        if ragam[0] in base_count:
            base_count[ragam[0]] += 1
        else:
            base_count[ragam[0]] = 1

    return base_count
```

```python
def calculate_base_probabilities(base_count):
        total = sum(base_count.values())

        base_probabilities = {}

        for note, count in base_count.items():
                base_probabilities[note] = count / total

        return base_probabilities
```

```python
def calculate_bigram_count(base_tuple):
    bigram_count = {}

    for ragam in base_tuple:
        for note_index in range(len(ragam) - 1):
            current_note = ragam[note_index]
            next_note = ragam[note_index + 1]

            if current_note not in bigram_count:
                bigram_count[current_note] = {}

            if next_note not in bigram_count[current_note]:
                bigram_count[current_note][next_note] = 1
            else:
                bigram_count[current_note][next_note] += 1

    return bigram_count
```

```python
def calculate_bigram_probabilities(bigram_count):
    bigram_probabilities = {}

    for note, next_dict in bigram_count.items():
        total_count = sum(next_dict.values())

        bigram_probabilities[note] = {}

        for next_note, count in next_dict.items():
            bigram_probabilities[note][next_note] = count / total_count

    return bigram_probabilities
```

```python
base_tuple_aarohanam = tuple(df_unique['Aarohanam'])

base_count_aarohanam = calculate_base_count(base_tuple_aarohanam)
base_probabilities_aarohanam = calculate_base_probabilities(base_count_aarohanam)

bigram_count_aarohanam = calculate_bigram_count(base_tuple_aarohanam)
bigram_probabilities_aarohanam = calculate_bigram_probabilities(bigram_count_aarohanam)
```

```python
def calculate_rarity(notes_list, base_probabilities, bigram_probabilities):
    total_score = base_probabilities[notes_list[0]]

    for i in range(len(notes_list) - 1):
        current_note = notes_list[i]
        next_note = notes_list[i + 1]

        if current_note in bigram_probabilities:
            next_dict = bigram_probabilities[current_note]
            if next_note in next_dict:
                score = next_dict[next_note]
                total_score *= score
            else:
                return 0
        else:
            return 0

    return total_score
```

```python
df_unique['multiplied_score_aarohanam'] = df_unique['Aarohanam'].apply(lambda x: calculate_rarity(x, base_probabilities_aarohanam, bigram_probabilities_aarohanam))
df_unique['multiplied_score_avarohanam'] = df_unique['Avarohanam'].apply(lambda x: calculate_rarity(x, base_probabilities_avarohanam, bigram_probabilities_avarohanam))

df_unique
```

```python
df_unique['multiplied_score'] = (df_unique['multiplied_score_aarohanam']*df_unique['multiplied_score_avarohanam'])**(0.5)

df_unique['multiplied_score']
```

### Perplexity-based Rarity Score

```python
def calculate_perplexity(notes_list, base_probabilities, bigram_probabilities):
    total_score = base_probabilities[notes_list[0]]
    list_length = len(notes_list)

    for i in range(list_length - 1):
        current_note = notes_list[i]
        next_note = notes_list[i + 1]

        if current_note in bigram_probabilities:
            next_dict = bigram_probabilities[current_note]
            if next_note in next_dict:
                score = next_dict[next_note]
                total_score *= score
            else:
                return 0
        else:
            return 0

    return (total_score)**(1/list_length)
```

```python
df_unique['multiplied_perplexity_aarohanam'] = df_unique['Aarohanam'].apply(lambda x: calculate_perplexity(x, base_probabilities_aarohanam, bigram_probabilities_aarohanam))
df_unique['multiplied_perplexity_avarohanam'] = df_unique['Avarohanam'].apply(lambda x: calculate_perplexity(x, base_probabilities_avarohanam, bigram_probabilities_avarohanam))

df_unique
```

```python
df_unique['multiplied_perplexity'] = (df_unique['multiplied_perplexity_aarohanam']*df_unique['multiplied_perplexity_avarohanam'])**(0.5)

df_unique['multiplied_perplexity']
```
